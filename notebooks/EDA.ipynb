```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt  
import seaborn as sns
```

```python
file_path = r"C:\Users\shasw\OneDrive\Desktop\EEG-Neuromarkiting-Project\combined_text_files.csv"
data = pd.read_csv(file_path)
```

```python
missing_values = data.isnull().sum()
print("Missing Values in Dataset:\n", missing_values)
```

```python
print("Basic Statistics:\n", data.describe())
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='label', data=data, palette='coolwarm')
plt.title("Label Distribution (Like vs. Dislike)")
plt.show()
```

```python
sample_user = data[data['name'] == 'Abhishek']  # Change name if needed
sample_product = sample_user[sample_user['product'] == 1]
plt.figure(figsize=(12, 6))
for i in range(2, 16):  # EEG features start from column index 2
    plt.plot(sample_product.iloc[:, i], label=f"Feature {i-2}")
plt.title("EEG Signals for User: Abhishek, Product: 1")
plt.xlabel("Time Points")
plt.ylabel("EEG Signal Value")
plt.legend()
plt.show()
```

```python
persons = ["Abhishek", "Ankur", "Gautam", "Girvar"]  # Replace with actual names
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle("EEG Signal Comparison for Product 1 Across 4 People")

for idx, person in enumerate(persons):
    row, col = divmod(idx, 2)
    person_data = data[(data['name'] == person) & (data['product'] == 1)]
    axes[row, col].plot(person_data.iloc[:, 2], label=person)  # Plot first EEG channel as an example
    axes[row, col].set_title(f"EEG for {person}")
    axes[row, col].set_xlabel("Time Points")
    axes[row, col].set_ylabel("EEG Signal Value")
    axes[row, col].legend()

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
